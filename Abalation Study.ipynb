# Abalation Study

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
from gensim.test.utils import common_corpus, common_dictionary
from gensim.models.coherencemodel import CoherenceModel
from nltk.tokenize import word_tokenize
import gensim.corpora as corpora
```

```python
cwd = os.getcwd()
```

```python
import gensim
def get_best_config(run_name):
    data_dir = os.path.join(cwd,"data\\wcai\\"+run_name)
    results_csv = os.path.join(data_dir,"results.csv")
    df = pd.read_csv(results_csv,usecols=["id","train_auc","test_auc","fcm.lam","fcm.rho","fcm.eta"])
    
    
    coh = []
    for index,row in df.iterrows():
        folder_name = row["id"]
        concept_path = os.path.join(data_dir,folder_name+"/concept/")
        words = []
        for i in range(0,30):
#             print("###################")
            file_path = concept_path + "epoch" + str(i) + ".txt" 
            file = open(file_path, 'r')
            Lines = file.readlines()
            topics = []
            for line in Lines:
                l = line.split(":")[1].strip("\n")
                topics+=(word_tokenize(l))
            words.append(topics)
        dictionary = corpora.Dictionary(words)
        doc_term_matrix = [dictionary.doc2bow(doc) for doc in words]
    
        Lda = gensim.models.ldamodel.LdaModel
        ldamodel = Lda(doc_term_matrix, num_topics=2, id2word = dictionary, passes=50)
        cm = CoherenceModel(model=ldamodel, corpus=common_corpus, coherence='u_mass')
        coherence = cm.get_coherence()
        coh.append(coherence)
    df["coherence"] = coh
    return df
    
```
